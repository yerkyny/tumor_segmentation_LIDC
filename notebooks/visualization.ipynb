```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
import nibabel as nib
import cv2
```

# Visualization and shapes

```python
ct_path = "/home/yerkyn/tumor_seg/data/img/LIDC-IDRI-0001-1-1.nii.gz"
mask_path = "/home/yerkyn/tumor_seg/data/seg/LIDC-IDRI-0001-1-1.nii.gz"

ct_img = nib.load(ct_path)
mask_img = nib.load(mask_path)

ct_data = ct_img.get_fdata()
mask_data = mask_img.get_fdata()

print("CT data shape:", ct_data.shape)
print("Mask data shape:", mask_data.shape)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(ct_data[:, :, ct_data.shape[2] // 2], cmap='gray')
plt.title('CT Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(mask_data[:, :, mask_data.shape[2] // 2])
plt.title('Mask')
plt.axis('off')

plt.show()
```

# Random Noise

```python
img = nib.load(ct_path)  
data = img.get_fdata()

slice_index = np.random.randint(data.shape[0])
slice_data = data[slice_index]
noise = np.random.normal(loc=0, scale=20, size=slice_data.shape)
noisy_slice = slice_data + noise
noisy_slice = np.clip(noisy_slice, -1024, 300)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(slice_data, cmap='gray')
ax[0].set_title('Original CT Slice')
ax[1].imshow(noisy_slice, cmap='gray')
ax[1].set_title('Noisy CT Slice')
plt.show()
```
